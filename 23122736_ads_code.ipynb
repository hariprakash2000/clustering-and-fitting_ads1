```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#read dataset
new_dataset = pd.read_csv('heart_failure_clinical_records_dataset.csv')


```

```python
# Calculate statistical moments and descriptive statistics
stats_summary = new_dataset.describe().T
stats_summary['mean'] = new_dataset.mean()
stats_summary['median'] = new_dataset.median()
stats_summary['std_dev'] = new_dataset.std()
stats_summary['skewness'] = new_dataset.skew()
stats_summary['kurtosis'] = new_dataset.kurtosis()

# Print the summary statistics
print("Statistical Summary:")
print(stats_summary)

# Calculate and visualize the correlation matrix
correlation_matrix = new_dataset.corr()
correlation_matrix
```

```python
# Relational Graph Function: Scatter Plot
def view_relational(data, x, y, title, xlabel, ylabel):
    """
    Creating a scatter plot to convey an xy relationship.


    """
    plt.figure(figsize=(8, 5))
    plt.scatter(data[x], data[y], alpha=0.75, edgecolor='black')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

view_relational(
    new_dataset, 
    x='age', 
    y='serum_creatinine', 
    title='Age vs Serum Creatinine',
    xlabel='Age',
    ylabel='Serum Creatinine (mg/dL)'
)
```

```python
# Categorical Graph Function: Bar Chart
def view_categorical(data, column, title, xlabel, ylabel):
    """
    Creates a bar chart to compare categories.

    """
    counts = data[column].value_counts()
    plt.figure(figsize=(8, 5))
    counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

view_categorical(
    new_dataset, 
    column='DEATH_EVENT', 
    title='Death Event Count',
    xlabel='Death Event (0 = Survived, 1 = Died)',
    ylabel='Count'
)

```

```python
# Statistical Graph Function: Heatmap
def view_statistical(matrix, title):
    """
    Creates a heatmap for statistical relationships.

    """
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(title)
    plt.show()

# Compute the correlation matrix
correlation_matrix = new_dataset.corr()
view_statistical(correlation_matrix, title='Correlation Heatmap')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import numpy as np

def kmeans_clustering(data, features, n_clusters):
    """
    Performs k-means clustering with normalization and silhouette evaluation for dataset.

    
    """
    # Select features for clustering
    clustering_data = data[features].dropna()
    
    # Normalize data
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(clustering_data)
    
    # Fit k-means
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(scaled_data)
    silhouette = silhouette_score(scaled_data, labels)
    
    # Back-scale cluster centers
    centers = scaler.inverse_transform(kmeans.cluster_centers_)
    
    # Print silhouette score
    print(f"Silhouette Score: {silhouette:.2f}")
    
    # Plot clusters
    plt.figure(figsize=(8, 5))
    plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=labels, cmap='viridis', alpha=0.6, label='Data Points')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
    plt.title('Clustered Data')
    plt.xlabel(features[0])
    plt.ylabel(features[1])
    plt.legend()
    plt.show()
    
    return kmeans, labels, centers

# Perform clustering on selected features
kmeans_model, cluster_labels, cluster_centers = kmeans_clustering(
    new_dataset,
    features=['age', 'serum_creatinine'],
    n_clusters=3
)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Perform clustering
def perform_kmeans_clustering(data, features, n_clusters):
    """
    Performs k-means clustering and returns the fitted model, labels, and scaler.

        
    """
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(data[features])
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(scaled_data)
    return kmeans, scaler, kmeans.labels_

# Define features for clustering
features = ['age', 'serum_creatinine']
kmeans_model, scaler, cluster_labels = perform_kmeans_clustering(new_dataset, features, n_clusters=3)

# Predict cluster membership and plot
def predict_cluster_membership_and_plot(kmeans, scaler, data, new_points, features):
    """
    Predicts cluster membership for new data points and visualizes them on the clustering plot.

    
    """
    # Normalize new points
    scaled_points = scaler.transform(new_points)
    predictions = kmeans.predict(scaled_points)
    
    # Normalize original data for consistent plotting
    scaled_data = scaler.transform(data[features])
    
    # Plot original clusters
    plt.figure(figsize=(8, 5))
    plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.6, label='Clustered Data')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
    
    # Plot new points
    plt.scatter(scaled_points[:, 0], scaled_points[:, 1], c='blue', edgecolor='black', s=100, label='predicted Points')
    
    # Titles and labels
    plt.title('Clustered Data with Predictions')
    plt.xlabel(features[0])
    plt.ylabel(features[1])
    plt.legend()
    plt.show()

    return predictions

# Define new points for prediction
pred_points = np.array([[60, 1.5], [45, 1.0], [70, 2.0]])  # Example new points

# Perform predictions and visualize
cluster_predictions = predict_cluster_membership_and_plot(
    kmeans_model,
    scaler,
    new_dataset,
    pred_points,
    features=features
)

# Output predictions
print("Cluster Predictions for New Points:", cluster_predictions)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

def fit_and_plot(data, x, y, new_values):
    """
    Fits a linear regression model, plots the fit with confidence intervals,
    and includes predictions for new values on the same plot.

    
    """
    # Prepare data
    X = data[[x]].values
    y_values = data[y].values
    
    # Fit regression model
    model = LinearRegression()
    model.fit(X, y_values)
    y_pred = model.predict(X)
    
    # Calculate residuals and confidence interval
    residuals = y_values - y_pred
    std_error = np.std(residuals)
    confidence_interval = 1.96 * std_error  # 95% confidence interval
    
    # Predictions for new values
    new_values = np.array(new_values).reshape(-1, 1)
    predictions = model.predict(new_values)
    uncertainty = 1.96 * std_error  
    
    plt.figure(figsize=(8, 5))
    plt.scatter(X, y_values, alpha=0.75, label='Data Points')
    plt.plot(X, y_pred, color='red', label='Regression Line')
    plt.fill_between(X.flatten(), y_pred - confidence_interval, y_pred + confidence_interval,
                     color='red', alpha=0.2, label='Confidence Interval')
    plt.errorbar(new_values.flatten(), predictions, yerr=uncertainty, fmt='o', capsize=5, color='blue', label='Predicted Points')
    plt.title('Linear Regression Fit with Predictions')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend()
    plt.show()

    return predictions, uncertainty

# Fit regression, plot, and predict new values
predicted_val = [30, 50, 70]  # Example new ages
predicted_serum_creatinine, uncertainty = fit_and_plot(
    new_dataset,
    x='age',
    y='serum_creatinine',
    new_values=predicted_val
)

# Output predictions and uncertainty
print("Predicted Serum Creatinine:", predicted_serum_creatinine)
print("Uncertainty:", uncertainty)
```
